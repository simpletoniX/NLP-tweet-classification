```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from collections import Counter
```

```python
df = pd.read_csv("/Users/tonix/PycharmProjects/pythonProject/Cyberbulling/cyberbullying_tweets.csv")
df.head()
```

```python
df.describe()
```

```python
# Types of bullying
df_types = set(df['cyberbullying_type'])
df_types
```

```python
# Create a plot of cyberbullying distribution
df_distribution = Counter(df['cyberbullying_type'])
df_distribution 
```

```python
fig, ax = plt.subplots()
ax.pie(df_distribution.values(), labels=df_distribution.keys(), autopct='%1.1f%%')
plt.title('Distribution', fontsize=18, color='green')
```

```python
# Max and min count of words
word_count = df['tweet_text'].str.split().str.len()
df['word_count'] = word_count

word_count_max100 = df.nlargest(100, 'word_count').to_csv('Max100.csv')
```

```python
# Min 
pd.set_option('display.max_rows', 20)

word_count_min100 = df.nsmallest(100, 'word_count').to_csv('Min100.csv')
```

```python
plt.figure(figsize=(6, 3.5))

plt.bar(df['cyberbullying_type'], df['word_count'], width=0.4)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.title("Count of words", fontsize=18, color='green')
plt.xlabel('Type', fontsize=12, labelpad=20)
plt.ylabel('Word', fontsize=12, labelpad=20)

plt.tick_params(which='major', pad=10)
```
